```python
path = "/api/events/"
base_url = "https://analystic-api-production.up.railway.app"
create_endpoint = f"{base_url}{path}"

response = requests.post(create_endpoint, json={"page": "/test+", "description": '123'}) # HTTP POST
print(response.headers)
print('ok', response.ok)
if response.ok:
  data = response.json()
  print(data, type(data), data.get('items'))
else:
  print(response.text)
```

```python
detail_path = "/api/events/1/"
detail_endpoint = f"{base_url}{detail_path}"
headers = {
  "Content-Type": "application/json"
}
response = requests.put(detail_endpoint, json={"description": "Inline test"}) # HTTP PUT
print(response.headers)
print('ok', response.status_code)
if response.ok:
  data = response.json()
  print(data, type(data), data.get('items'))
else:
  print(response.text)
```

```python
from faker import Faker
import random
events = 500
pages = [
  "/", "/about", "/pricing", "/contact",
  "/blog", "/products", "/login", "/singup",
  "/dashboard", "/settings"
]
fake = Faker()
session_ids = [fake.uuid4() for _ in range(20)]
path = "/api/events/"
base_url = "http://localhost:8002"
base_url = "https://analystic-api-production.up.railway.app"
create_endpoint = f"{base_url}{path}"
referrers = [
  "https://google.com",
  "https://facebook.com",
  "https://twitter.com",
  "https://linkedin.com",
  "",
  "https://github.com"
]

for i in range(events):
    user_agent = random.choice([
        fake.chrome,
        fake.firefox,
        fake.safari,
        fake.opera,
        fake.user_agent
    ])()

    payload = dict(
        page=random.choice(pages),
        user_agent=user_agent,
        ip_address=fake.ipv4(),
        referrer=random.choice(referrers),
        session_id=random.choice(session_ids),
        duration=str(random.randint(50, 5000)) # Transformado em String
    )

    response = requests.post(create_endpoint, json=payload)
    
    if response.ok:
        data = response.json()
        print(data)
```
